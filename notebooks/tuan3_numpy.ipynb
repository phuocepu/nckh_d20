**Váº¥n Ä‘á»:** TÃ­nh tá»•ng 1 + 2 + 3 + ... + 10,000,000

Python cÃ³ vÃ²ng láº·p `for`, nhÆ°ng khi dá»¯ liá»‡u lá»›n â†’ **quÃ¡ cháº­m!**

````python
# === DEMO: CUá»˜C ÄUA Tá»C Äá»˜ ===
import time
import numpy as np

n = 10_000_000  # 10 triá»‡u

# CÃ¡ch 1: VÃ²ng láº·p for
start = time.time()
tong_for = 0
for i in range(n):
    tong_for += i
time_for = time.time() - start

# CÃ¡ch 2: NumPy
start = time.time()
tong_np = np.sum(np.arange(n))
time_np = time.time() - start

print(f"ğŸŒ For loop: {time_for:.4f} giÃ¢y")
print(f"âš¡ NumPy:    {time_np:.4f} giÃ¢y")
print(f"\nğŸ† NumPy nhanh hÆ¡n {time_for/time_np:.0f} láº§n!")
print(f"\nKáº¿t quáº£ giá»‘ng nhau: {tong_for == tong_np}")
````

### ğŸ¬ á»¨NG Dá»¤NG THá»°C Táº¾: Speed Race Animation

**Kiáº¿n thá»©c sá»­ dá»¥ng:** So sÃ¡nh tá»‘c Ä‘á»™ for loop vs NumPy

Xem cuá»™c Ä‘ua tá»‘c Ä‘á»™ xá»­ lÃ½ dá»¯ liá»‡u!

````python
# ğŸ¬ DEMO: Speed Race Animation
# Cháº¡y cell nÃ y Ä‘á»ƒ xem hiá»‡u á»©ng!

import time
from IPython.display import clear_output
import numpy as np

print("ğŸ CUá»˜C ÄUA Tá»C Äá»˜: For Loop vs NumPy")
print("=" * 45)
time.sleep(1)

# CÃ¡c kÃ­ch thÆ°á»›c dá»¯ liá»‡u tÄƒng dáº§n
sizes = [1000, 10000, 100000, 500000, 1000000, 5000000]

for n in sizes:
    clear_output(wait=True)
    print("ğŸ CUá»˜C ÄUA Tá»C Äá»˜: For Loop vs NumPy")
    print("=" * 45)
    print()
    print(f"ğŸ“Š Dá»¯ liá»‡u: {n:,} sá»‘")
    print()
    
    # Äo for loop
    start = time.time()
    s = 0
    for i in range(n):
        s += i
    t_for = time.time() - start
    
    # Äo NumPy
    start = time.time()
    s = np.sum(np.arange(n))
    t_np = time.time() - start
    
    # Váº½ thanh so sÃ¡nh
    max_bar = 30
    if t_for > 0:
        bar_for = min(max_bar, int(t_for / max(t_for, 0.001) * max_bar))
        bar_np = max(1, int(t_np / max(t_for, 0.001) * max_bar))
    else:
        bar_for = max_bar
        bar_np = 1
    
    print(f"ğŸŒ For loop: {'â–ˆ' * bar_for}{'â–‘' * (max_bar - bar_for)} {t_for:.4f}s")
    print(f"âš¡ NumPy:    {'â–ˆ' * bar_np}{'â–‘' * (max_bar - bar_np)} {t_np:.4f}s")
    print()
    
    if t_np > 0:
        print(f"ğŸ† NumPy nhanh hÆ¡n {t_for/t_np:.0f}x")
    
    time.sleep(1.5)

clear_output(wait=True)
print("ğŸ CUá»˜C ÄUA Tá»C Äá»˜: Káº¾T QUáº¢")
print("=" * 45)
print()
print("âœ… NumPy LUÃ”N tháº¯ng!")
print("ğŸ“ˆ Dá»¯ liá»‡u cÃ ng lá»›n â†’ chÃªnh lá»‡ch cÃ ng nhiá»u")
print()
print("ğŸ’¡ Kiáº¿n thá»©c: NumPy xá»­ lÃ½ array cÃ¹ng lÃºc (vectorized)")
print("   thay vÃ¬ tá»«ng pháº§n tá»­ má»™t (for loop)")
````

### ğŸ¬ á»¨NG Dá»¤NG THá»°C Táº¾: Physics Simulator

**Kiáº¿n thá»©c sá»­ dá»¥ng:** PhÃ©p tÃ­nh vector hÃ³a Ä‘á»ƒ tÃ­nh quá»¹ Ä‘áº¡o bay

MÃ´ phá»ng quá»¹ Ä‘áº¡o nÃ©m xiÃªn báº±ng NumPy!

````python
# ğŸ¬ DEMO: Physics Simulator - NÃ©m xiÃªn
# Cháº¡y cell nÃ y Ä‘á»ƒ xem hiá»‡u á»©ng!

import time
from IPython.display import clear_output
import numpy as np

# ThÃ´ng sá»‘ ban Ä‘áº§u
v0 = 20       # váº­n tá»‘c ban Ä‘áº§u (m/s)
g = 9.8       # gia tá»‘c trá»ng trÆ°á»ng
goc_do = 45   # gÃ³c nÃ©m (Ä‘á»™)

# Chuyá»ƒn sang radian (vector hÃ³a!)
goc_rad = goc_do * np.pi / 180

# TÃ­nh thá»i gian bay
t_max = 2 * v0 * np.sin(goc_rad) / g

# Táº¡o array thá»i gian
t = np.linspace(0, t_max, 50)

# Vector hÃ³a: TÃ­nh x, y cho Táº¤T Cáº¢ thá»i Ä‘iá»ƒm cÃ¹ng lÃºc!
x = v0 * np.cos(goc_rad) * t
y = v0 * np.sin(goc_rad) * t - 0.5 * g * t**2
y = np.maximum(y, 0)  # KhÃ´ng cho y < 0

# Animation
chieu_rong = 50
chieu_cao = 15
x_max = np.max(x)
y_max = np.max(y)

for i in range(len(t)):
    clear_output(wait=True)
    print(f"ğŸ¯ MÃ” PHá»NG NÃ‰M XIÃŠN (gÃ³c {goc_do}Â°, vâ‚€ = {v0} m/s)")
    print("=" * 55)
    
    # Chuyá»ƒn tá»a Ä‘á»™ thá»±c sang tá»a Ä‘á»™ mÃ n hÃ¬nh
    col = int(x[i] / max(x_max, 1) * (chieu_rong - 1))
    row = chieu_cao - 1 - int(y[i] / max(y_max, 1) * (chieu_cao - 2))
    
    for r in range(chieu_cao):
        dong = ""
        for c in range(chieu_rong):
            if r == row and c == col:
                dong += "ğŸ”´"
            elif r == chieu_cao - 1:
                dong += "â–¬â–¬"
            else:
                dong += "  "
        print(dong)
    
    print(f"\nğŸ“ x = {x[i]:.1f}m  |  y = {y[i]:.1f}m  |  t = {t[i]:.2f}s")
    time.sleep(0.1)

print(f"\nâœ… Táº§m bay: {x_max:.1f}m | Äá»™ cao max: {y_max:.1f}m")
print(f"\nğŸ’¡ Kiáº¿n thá»©c: Vector hÃ³a tÃ­nh x, y cho 50 thá»i Ä‘iá»ƒm trong 1 dÃ²ng code!")
print(f"   x = vâ‚€Â·cos(Î¸)Â·t      â†’  x = v0 * np.cos(goc_rad) * t")
print(f"   y = vâ‚€Â·sin(Î¸)Â·t - Â½gtÂ²  â†’  y = v0 * np.sin(goc_rad) * t - 0.5 * g * t**2")
````

### ğŸ¬ á»¨NG Dá»¤NG THá»°C Táº¾: Matrix Rain

````python
# ğŸ¬ DEMO: Matrix Rain Animation
# Cháº¡y cell nÃ y Ä‘á»ƒ xem hiá»‡u á»©ng!

import time
from IPython.display import clear_output
import numpy as np

rows, cols = 15, 30

# Táº¡o "giá»t mÆ°a" - vá»‹ trÃ­ cá»™t vÃ  hÃ ng hiá»‡n táº¡i
so_giot = 8
cot_giot = np.random.randint(0, cols, so_giot)
hang_giot = np.random.randint(0, rows, so_giot)
ky_tu = "0123456789ABCDEF"

for frame in range(40):
    clear_output(wait=True)
    print("ğŸ’š MATRIX RAIN (Array 2D)")
    print("=" * 62)
    
    # Táº¡o matrix 2D toÃ n khoáº£ng tráº¯ng
    matrix = np.full((rows, cols), " ", dtype="<U1")
    
    # Äáº·t kÃ½ tá»± random táº¡i vá»‹ trÃ­ giá»t mÆ°a
    for g in range(so_giot):
        # Vá»‡t Ä‘uÃ´i (3 kÃ½ tá»± phÃ­a trÃªn)
        for trail in range(4):
            r = (hang_giot[g] - trail) % rows
            matrix[r, cot_giot[g]] = ky_tu[np.random.randint(0, len(ky_tu))]
    
    # In matrix
    for r in range(rows):
        dong = " ".join(matrix[r])
        print(f"  {dong}")
    
    # Di chuyá»ƒn giá»t mÆ°a xuá»‘ng (vector hÃ³a!)
    hang_giot = (hang_giot + 1) % rows
    
    # Random Ä‘á»•i cá»™t
    doi_cot = np.random.random(so_giot) < 0.1
    cot_giot[doi_cot] = np.random.randint(0, cols, np.sum(doi_cot))
    
    print(f"\n  Shape: ({rows}, {cols}) | Frame: {frame+1}/40")
    time.sleep(0.15)

clear_output(wait=True)
print("ğŸ’š MATRIX RAIN - HOÃ€N THÃ€NH")
print("=" * 62)
print()
print("âœ… Array 2D táº¡o ra hiá»‡u á»©ng áº¥n tÆ°á»£ng!")
print()
print("ğŸ’¡ Kiáº¿n thá»©c:")
print("   â€¢ np.full((rows, cols), ' ') â†’ táº¡o matrix 2D")
print("   â€¢ matrix[r, c] = value       â†’ gÃ¡n giÃ¡ trá»‹")
print("   â€¢ Vector hÃ³a di chuyá»ƒn: hang = (hang + 1) % rows")
````

# ğŸ¬ DEMO CUá»I: Monte Carlo tÃ­nh sá»‘ Pi

**Ã tÆ°á»Ÿng:**
1. HÃ¬nh vuÃ´ng cáº¡nh 1, bÃªn trong cÃ³ 1/4 hÃ¬nh trÃ²n bÃ¡n kÃ­nh 1
2. "NÃ©m" random hÃ ng triá»‡u Ä‘iá»ƒm vÃ o hÃ¬nh vuÃ´ng
3. Äáº¿m Ä‘iá»ƒm rÆ¡i trong hÃ¬nh trÃ²n (xÂ² + yÂ² â‰¤ 1)
4. **Ï€/4 â‰ˆ Ä‘iá»ƒm_trong_trÃ²n / tá»•ng_Ä‘iá»ƒm**

**Kiáº¿n thá»©c NumPy:** random, vector hÃ³a (xÂ²+yÂ²), boolean indexing, np.sum()

ğŸ“Œ **Slide 19**

````python
# ğŸ¬ DEMO: Monte Carlo tÃ­nh Pi
# Cháº¡y cell nÃ y Ä‘á»ƒ xem hiá»‡u á»©ng!

import time
from IPython.display import clear_output
import numpy as np

print("ğŸ¯ MONTE CARLO TÃNH Sá» PI")
print("=" * 50)
time.sleep(1)

# CÃ¡c bÆ°á»›c tÄƒng dáº§n sá»‘ Ä‘iá»ƒm
so_diem_list = [100, 500, 1000, 5000, 10000, 50000, 100000, 500000, 1000000]

for n in so_diem_list:
    clear_output(wait=True)
    print("ğŸ¯ MONTE CARLO TÃNH Sá» PI")
    print("=" * 50)
    print()
    
    # Sinh random n Ä‘iá»ƒm (x, y) trong [0, 1]
    x = np.random.random(n)
    y = np.random.random(n)
    
    # TÃ­nh khoáº£ng cÃ¡ch (vector hÃ³a!)
    khoang_cach = x**2 + y**2
    
    # Äáº¿m Ä‘iá»ƒm trong hÃ¬nh trÃ²n (boolean indexing!)
    trong_tron = np.sum(khoang_cach <= 1)
    
    # TÃ­nh Pi
    pi_uoc_luong = 4 * trong_tron / n
    sai_so = abs(pi_uoc_luong - np.pi)
    
    print(f"ğŸ“Š Sá»‘ Ä‘iá»ƒm: {n:>10,}")
    print(f"   Trong trÃ²n: {trong_tron:>10,}")
    print()
    
    # Váº½ hÃ¬nh Ä‘Æ¡n giáº£n (grid 20x20)
    if n <= 5000:
        grid_size = 20
        for r in range(grid_size):
            dong = "   "
            for c in range(grid_size):
                cx = c / grid_size
                cy = 1 - r / grid_size
                # Kiá»ƒm tra cÃ³ Ä‘iá»ƒm nÃ o gáº§n vá»‹ trÃ­ nÃ y khÃ´ng
                nearby = np.sum((np.abs(x - cx) < 0.03) & (np.abs(y - cy) < 0.03))
                if cx**2 + cy**2 <= 1:
                    dong += "ğŸŸ¢" if nearby > 0 else "Â· "
                else:
                    dong += "ğŸ”´" if nearby > 0 else "  "
            print(dong)
        print()
    
    # Hiá»ƒn thá»‹ káº¿t quáº£
    print("â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—")
    print(f"â•‘  Ï€ â‰ˆ {pi_uoc_luong:.6f}                         â•‘")
    print(f"â•‘  Ï€ =  {np.pi:.6f} (thá»±c táº¿)                â•‘")
    print(f"â•‘  Sai sá»‘: {sai_so:.6f}                      â•‘")
    print("â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•")
    
    # Thanh chÃ­nh xÃ¡c
    do_chinh_xac = max(0, 100 - sai_so * 1000)
    bar_len = int(do_chinh_xac / 5)
    print(f"\n   Äá»™ chÃ­nh xÃ¡c: [{'â–ˆ' * bar_len}{'â–‘' * (20 - bar_len)}] {do_chinh_xac:.1f}%")
    
    time.sleep(1.5)

clear_output(wait=True)
print("ğŸ¯ MONTE CARLO TÃNH Sá» PI - Káº¾T QUáº¢")
print("=" * 50)
print()
print(f"âœ… Vá»›i 1,000,000 Ä‘iá»ƒm: Ï€ â‰ˆ {pi_uoc_luong:.6f}")
print(f"   Sai sá»‘ chá»‰: {sai_so:.6f}")
print()
print("ğŸ’¡ Kiáº¿n thá»©c NumPy Ä‘Ã£ dÃ¹ng:")
print("   â€¢ np.random.random(n)   â†’ sinh n sá»‘ random")
print("   â€¢ x**2 + y**2           â†’ vector hÃ³a")
print("   â€¢ np.sum(mask)          â†’ Ä‘áº¿m True (boolean indexing)")
print("   â€¢ CÃ ng nhiá»u Ä‘iá»ƒm â†’ cÃ ng chÃ­nh xÃ¡c!")
````
