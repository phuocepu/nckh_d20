# Section 9: Lỗi thường gặp với NumPy

📌 **Slide 11**

| Lỗi | Nguyên nhân | Cách phòng |
|-----|------------|------------|
| ValueError (shape) | Mảng khác kích thước | Kiểm tra `.shape` |
| LinAlgError | Ma trận suy biến | Kiểm tra `det != 0` |
| AxisError | Axis không tồn tại | Kiểm tra `.ndim` |

````python
# === CODE LỖI: NumPy ===

# Lỗi 1: Shape mismatch
a = np.array([1, 2, 3])
b = np.array([1, 2])
print(f"a.shape = {a.shape}, b.shape = {b.shape}")
c = a + b   # ValueError!
````

````python
# === BÀI TẬP: Sửa 3 lỗi NumPy ===

# Lỗi 1: Shape mismatch khi cộng
a = np.array([1, 2, 3])
b = np.array([10, 20])
c = a + b   # ... sửa b để cùng shape với a
print(c)

# Lỗi 2: Ma trận suy biến (không thể nghịch đảo)
A = np.array([[1, 2], [2, 4]])   # det = 0!
A_inv = np.linalg.inv(A)   # ... sửa ma trận A
print(A_inv)

# Lỗi 3: AxisError
x = np.array([1, 2, 3])   # 1D array
print(np.sum(x, axis=1))   # ... sửa axis
````

# Section 12: Print Debugging

📌 **Slide 14**

Kỹ thuật đơn giản nhất: thêm `print()` để "nhìn thấy" giá trị biến qua từng bước.

````python
# === CODE MẪU: Print Debugging ===
# Hàm tính trung bình bị SAI - dùng print để tìm bug

def trung_binh(ds):
    tong = 0
    for i in range(1, len(ds)):   # BUG ở đây!
        tong += ds[i]
        print(f"  i={i}, ds[i]={ds[i]}, tong={tong}")  # ← debug
    tb = tong / len(ds)
    print(f"  tong={tong}, len={len(ds)}, tb={tb}")  # ← debug
    return tb

data = [10, 20, 30]
print(f"Kết quả: {trung_binh(data)}")
print(f"Đáp án đúng: {sum(data)/len(data)}")
````

````python
# === BÀI TẬP: Dùng print debugging để tìm bug ===
# Hàm đếm số nguyên tố nhỏ hơn n bị SAI
# Thêm print() vào để tìm bug, rồi sửa

def dem_nguyen_to(n):
    dem = 0
    for so in range(2, n):
        la_nguyen_to = True
        for i in range(2, so):
            if so % i == 0:
                la_nguyen_to = True   # BUG: ???
                break
        if la_nguyen_to:
            dem += 1
    return dem

print(f"Số nguyên tố < 20: {dem_nguyen_to(20)}")
print("Đáp án đúng: 8 (2,3,5,7,11,13,17,19)")
````

# Section 13: Try - Except

📌 **Slide 15**

Bắt lỗi có kiểm soát - chương trình không crash.

```python
try:
    # Code có thể lỗi
except LoaiLoi as e:
    # Xử lý lỗi
finally:
    # Luôn chạy (tùy chọn)
```

````python
# === BÀI 1: Đoán output TRƯỚC KHI CHẠY ===
# Viết đáp án vào comment, rồi mới chạy kiểm tra

# Đáp án dự đoán: ...

def bi_an(n):
    if n <= 1:
        return n
    return bi_an(n - 1) + bi_an(n - 2)

ket_qua = [bi_an(i) for i in range(8)]
print(ket_qua)
````

````python
# === BÀI 2: Đoán output TRƯỚC KHI CHẠY ===

# Đáp án dự đoán: ...

def xu_ly(ds):
    ket_qua = []
    for x in ds:
        if x % 2 == 0:
            ket_qua.append(x ** 2)
        else:
            ket_qua.append(x * 3 + 1)
    return ket_qua

print(xu_ly([1, 2, 3, 4, 5]))
````

````python
# === BÀI 3: Đoán output TRƯỚC KHI CHẠY ===

# Đáp án dự đoán: ...

def ma_thuat(text):
    vowels = 'aeiou'
    result = ''
    for ch in text.lower():
        if ch in vowels:
            result += ch.upper()
        else:
            result += ch
    return result

print(ma_thuat("Hello World"))
````

````python
# === TÌM VÀ SỬA 5 LỖI ===
# Copy cell này, sửa hết 5 lỗi, chạy thành công!

import numpy as np
import matplotlib.pyplt as plt            # 🔍 Lỗi 1
import pandas as pd

df = pd.read_csv('diem_sinh_vien.csv')

# Vẽ histogram điểm Toán
plt.figure(figsize=(10, 6))
plt.hist(df['toan'], bins=10)              # 🔍 Lỗi 2
plt.title('Phân phối điểm Toán')

# Vẽ scatter Toán vs Lý
x = df['Toan'].values
y = df['Ly'].values
colors = df['Hoa'].values

plt.figure(figsize=(8, 6))
plt.scatter(x, y, c=colors)
plt.xlabel('Toán')
plt.ylabel('Lý')
plt.title('Tương quan Toán - Lý')

# Tính trung bình và in
mon = ['Toan', 'Ly', 'Hoa', 'Anh']
for m in mon                               # 🔍 Lỗi 3
    tb = df[m].mean()
    print("TB " + m + ": " + tb)           # 🔍 Lỗi 4

# Vẽ bar chart trung bình
tb_list = [df[m].mean() for m in mon]
plt.figure(figsize=(8, 5))
plt.bar(mon, tb_list)
plt.title('Điểm trung bình các môn')
plt.savefig                                 # 🔍 Lỗi 5
plt.show()
````
