# 4. PH√âP SO S√ÅNH (Comparison Operators)

K·∫øt qu·∫£ lu√¥n l√† `True` ho·∫∑c `False`

| Ph√©p so s√°nh | K√Ω hi·ªáu |
|--------------|--------|
| B·∫±ng | == |
| Kh√°c | != |
| L·ªõn h∆°n | > |
| Nh·ªè h∆°n | < |
| L·ªõn h∆°n ho·∫∑c b·∫±ng | >= |
| Nh·ªè h∆°n ho·∫∑c b·∫±ng | <= |

````python
# === CODE M·∫™U ===
a = 10
b = 5

print(a == b)   # False (10 kh√¥ng b·∫±ng 5)
print(a != b)   # True  (10 kh√°c 5)
print(a > b)    # True  (10 l·ªõn h∆°n 5)
print(a < b)    # False (10 kh√¥ng nh·ªè h∆°n 5)
print(a >= 10)  # True  (10 >= 10)
print(a <= 5)   # False (10 kh√¥ng <= 5)
````
