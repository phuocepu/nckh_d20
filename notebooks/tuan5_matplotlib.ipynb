# Section 6: PIE CHART - Bi·ªÉu ƒë·ªì tr√≤n

üìå **Slide 8** | `plt.pie(sizes, labels=...)`

**Khi n√†o d√πng:** T·ª∑ l·ªá ph·∫ßn trƒÉm, ph√¢n b·ªï th√†nh ph·∫ßn

````python
# === CODE M·∫™U: S·ªë nguy√™n t·ªë vs h·ª£p s·ªë [1, 100] ===
def is_prime(n):
    if n < 2: return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0: return False
    return True

primes = sum(1 for n in range(1, 101) if is_prime(n))
composites = 100 - primes - 1  # tr·ª´ s·ªë 1

plt.figure(figsize=(8, 8))
plt.pie([primes, composites, 1],
        labels=[f'Nguy√™n t·ªë ({primes})', f'H·ª£p s·ªë ({composites})', 'S·ªë 1'],
        autopct='%1.1f%%', startangle=90,
        colors=['#e74c3c', '#3498db', '#95a5a6'],
        explode=(0.05, 0, 0))
plt.title('Ph√¢n lo·∫°i s·ªë trong [1, 100]')
plt.show()
````
