### üé¨ ·ª®NG D·ª§NG TH·ª∞C T·∫æ: Loading Bar

**Ki·∫øn th·ª©c s·ª≠ d·ª•ng:** Ph√©p s·ªë h·ªçc `+`, `/`, `*` v√† format string

T·∫°o thanh ti·∫øn tr√¨nh loading chuy√™n nghi·ªáp!

````python
# üé¨ DEMO: Loading Bar Animation
# Ch·∫°y cell n√†y ƒë·ªÉ xem hi·ªáu ·ª©ng!

import time
from IPython.display import clear_output

tong_buoc = 20
chieu_dai_bar = 30

print("üìä ƒêang t·∫£i d·ªØ li·ªáu...")
for i in range(tong_buoc + 1):
    clear_output(wait=True)
    
    # T√≠nh ph·∫ßn trƒÉm ho√†n th√†nh
    phan_tram = (i / tong_buoc) * 100
    
    # T√≠nh s·ªë k√Ω t·ª± ƒë√£ ho√†n th√†nh
    so_da_xong = int((i / tong_buoc) * chieu_dai_bar)
    so_chua_xong = chieu_dai_bar - so_da_xong
    
    # T·∫°o thanh progress bar
    thanh_bar = "‚ñà" * so_da_xong + "‚ñë" * so_chua_xong
    
    print(f"[{thanh_bar}] {phan_tram:.1f}%")
    time.sleep(0.1)

print("\n‚úÖ T·∫£i ho√†n t·∫•t!")
print("\nüí° Ki·∫øn th·ª©c: Ph√©p chia (/), nh√¢n (*), format string (:.1f)")
````

# 4. PH√âP SO S√ÅNH (Comparison Operators)

K·∫øt qu·∫£ lu√¥n l√† `True` ho·∫∑c `False`

### üé¨ ·ª®NG D·ª§NG TH·ª∞C T·∫æ: Password Strength Checker

**Ki·∫øn th·ª©c s·ª≠ d·ª•ng:** Ph√©p so s√°nh `>=`, `<`, `==` ƒë·ªÉ ki·ªÉm tra ƒë·ªô d√†i v√† ƒëi·ªÅu ki·ªán

T·∫°o c√¥ng c·ª• ki·ªÉm tra ƒë·ªô m·∫°nh m·∫≠t kh·∫©u nh∆∞ c√°c website chuy√™n nghi·ªáp!

````python
# üé¨ DEMO: Password Strength Checker
# Ch·∫°y cell n√†y ƒë·ªÉ xem hi·ªáu ·ª©ng!

import time
from IPython.display import clear_output

# Danh s√°ch m·∫≠t kh·∫©u ƒë·ªÉ test
passwords = ["abc", "hello123", "MyP@ssw0rd!", "admin", "SecurePass2024!"]

print("üîê PASSWORD STRENGTH CHECKER")
print("=" * 40)
print()

for pw in passwords:
    clear_output(wait=True)
    print("üîê PASSWORD STRENGTH CHECKER")
    print("=" * 40)
    print()
    print(f"ƒêang ki·ªÉm tra: {'*' * len(pw)}")
    print()
    time.sleep(0.5)
    
    # L·∫•y ƒë·ªô d√†i m·∫≠t kh·∫©u
    do_dai = len(pw)
    
    # Ki·ªÉm tra ƒë·ªô m·∫°nh b·∫±ng ph√©p SO S√ÅNH
    du_dai = do_dai >= 8           # ƒê·ªô d√†i >= 8?
    qua_ngan = do_dai < 6          # Qu√° ng·∫Øn < 6?
    ly_tuong = do_dai >= 12        # L√Ω t∆∞·ªüng >= 12?
    
    # T√≠nh ƒëi·ªÉm d·ª±a tr√™n so s√°nh
    diem = 0
    diem = diem + (du_dai == True) * 30      # +30 n·∫øu ƒë·ªß d√†i
    diem = diem + (ly_tuong == True) * 20    # +20 n·∫øu l√Ω t∆∞·ªüng
    diem = diem + (qua_ngan == False) * 20   # +20 n·∫øu kh√¥ng qu√° ng·∫Øn
    diem = diem + 30  # ƒêi·ªÉm c∆° b·∫£n
    
    # Hi·ªÉn th·ªã k·∫øt qu·∫£
    clear_output(wait=True)
    print("üîê PASSWORD STRENGTH CHECKER")
    print("=" * 40)
    print()
    print(f"M·∫≠t kh·∫©u: {pw}")
    print(f"ƒê·ªô d√†i: {do_dai} k√Ω t·ª±")
    print()
    print(f"‚îú‚îÄ ƒê·ªß d√†i (>= 8)?     {du_dai}")
    print(f"‚îú‚îÄ Qu√° ng·∫Øn (< 6)?   {qua_ngan}")
    print(f"‚îî‚îÄ L√Ω t∆∞·ªüng (>= 12)? {ly_tuong}")
    print()
    
    # Thanh ƒëi·ªÉm
    bar_len = diem // 5
    bar = "‚ñà" * bar_len + "‚ñë" * (20 - bar_len)
    
    # X·∫øp lo·∫°i
    if diem >= 80:
        xep_loai = "üü¢ M·∫†NH"
    elif diem >= 60:
        xep_loai = "üü° TRUNG B√åNH"
    else:
        xep_loai = "üî¥ Y·∫æU"
    
    print(f"[{bar}] {diem}%")
    print(f"ƒê√°nh gi√°: {xep_loai}")
    print()
    time.sleep(1.5)

print("‚úÖ Ki·ªÉm tra ho√†n t·∫•t!")
print("\nüí° Ki·∫øn th·ª©c: Ph√©p so s√°nh >=, <, == tr·∫£ v·ªÅ True/False")
````

````python
# üé¨ DEMO: Security Access Panel
# Ch·∫°y cell n√†y ƒë·ªÉ xem hi·ªáu ·ª©ng!

import time
from IPython.display import clear_output

# Danh s√°ch nh√¢n vi√™n v·ªõi c√°c ƒëi·ªÅu ki·ªán truy c·∫≠p
nhan_vien = [
    {"ten": "Nguy·ªÖn VƒÉn A", "the_hop_le": True, "van_tay": True, "la_admin": False},
    {"ten": "Tr·∫ßn Th·ªã B", "the_hop_le": True, "van_tay": False, "la_admin": False},
    {"ten": "L√™ VƒÉn C", "the_hop_le": False, "van_tay": True, "la_admin": False},
    {"ten": "Admin H·ªá Th·ªëng", "the_hop_le": False, "van_tay": False, "la_admin": True},
    {"ten": "Kh√°ch L·∫°", "the_hop_le": False, "van_tay": False, "la_admin": False},
]

print("üè¢ SECURITY ACCESS PANEL")
print("=" * 45)
print()
time.sleep(1)

for nv in nhan_vien:
    clear_output(wait=True)
    print("üè¢ SECURITY ACCESS PANEL")
    print("=" * 45)
    print()
    print(f"üîç ƒêang qu√©t: {nv['ten']}...")
    print()
    time.sleep(0.8)
    
    # L·∫•y th√¥ng tin
    the_hop_le = nv["the_hop_le"]
    van_tay = nv["van_tay"]
    la_admin = nv["la_admin"]
    
    # PH√âP LOGIC - Ki·ªÉm tra quy·ªÅn truy c·∫≠p
    # ƒêi·ªÅu ki·ªán 1: Th·∫ª h·ª£p l·ªá V√Ä v√¢n tay kh·ªõp (and)
    xac_thuc_2_lop = the_hop_le and van_tay
    
    # ƒêi·ªÅu ki·ªán 2: L√† admin HO·∫∂C ƒë√£ x√°c th·ª±c 2 l·ªõp (or)
    co_quyen_vao = la_admin or xac_thuc_2_lop
    
    # ƒêi·ªÅu ki·ªán 3: KH√îNG b·ªã kh√≥a (not)
    bi_khoa = not (the_hop_le or la_admin)  # Kh√≥a n·∫øu kh√¥ng c√≥ th·∫ª v√† kh√¥ng ph·∫£i admin
    
    clear_output(wait=True)
    print("üè¢ SECURITY ACCESS PANEL")
    print("=" * 45)
    print()
    print(f"üë§ Nh√¢n vi√™n: {nv['ten']}")
    print()
    print("üìã Ki·ªÉm tra ƒëi·ªÅu ki·ªán:")
    print(f"   ‚îú‚îÄ Th·∫ª h·ª£p l·ªá?        {the_hop_le}")
    print(f"   ‚îú‚îÄ V√¢n tay kh·ªõp?      {van_tay}")
    print(f"   ‚îî‚îÄ L√† Admin?          {la_admin}")
    print()
    print("üîê K·∫øt qu·∫£ ph√©p logic:")
    print(f"   ‚îú‚îÄ the AND van_tay  = {xac_thuc_2_lop}")
    print(f"   ‚îú‚îÄ admin OR xac_thuc = {co_quyen_vao}")
    print(f"   ‚îî‚îÄ NOT (the OR admin) = {bi_khoa}")
    print()
    
    # K·∫øt qu·∫£ cu·ªëi c√πng
    if bi_khoa:
        print("‚ïî‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïó")
        print("‚ïë  üö´ T·ª™ CH·ªêI - T√ÄI KHO·∫¢N B·ªä KH√ìA!     ‚ïë")
        print("‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù")
    elif co_quyen_vao:
        print("‚ïî‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïó")
        print("‚ïë  ‚úÖ CHO PH√âP TRUY C·∫¨P - C·ª¨A M·ªû!     ‚ïë")
        print("‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù")
    else:
        print("‚ïî‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïó")
        print("‚ïë  ‚ö†Ô∏è T·ª™ CH·ªêI - THI·∫æU X√ÅC TH·ª∞C!       ‚ïë")
        print("‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù")
    
    print()
    time.sleep(2)

clear_output(wait=True)
print("üè¢ SECURITY ACCESS PANEL")
print("=" * 45)
print()
print("‚úÖ Ho√†n th√†nh ki·ªÉm tra!")
print("\nüí° Ki·∫øn th·ª©c s·ª≠ d·ª•ng:")
print("   ‚Ä¢ and: C·∫£ hai ƒëi·ªÅu ki·ªán ph·∫£i True")
print("   ‚Ä¢ or: √çt nh·∫•t m·ªôt ƒëi·ªÅu ki·ªán True")
print("   ‚Ä¢ not: ƒê·∫£o ng∆∞·ª£c gi√° tr·ªã True/False")
````

### üé¨ ·ª®NG D·ª§NG TH·ª∞C T·∫æ: Bouncing Ball

**Ki·∫øn th·ª©c s·ª≠ d·ª•ng:** `if-else` ƒë·ªÉ ki·ªÉm tra ƒëi·ªÅu ki·ªán bi√™n

T·∫°o hi·ªáu ·ª©ng qu·∫£ b√≥ng n·∫£y trong khung!

````python
# üé¨ DEMO: Bouncing Ball Animation
# Ch·∫°y cell n√†y ƒë·ªÉ xem hi·ªáu ·ª©ng!

import time
from IPython.display import clear_output

# K√≠ch th∆∞·ªõc khung
chieu_rong = 30
chieu_cao = 10

# V·ªã tr√≠ ban ƒë·∫ßu
x = 0
y = 0

# H∆∞·ªõng di chuy·ªÉn
dx = 1  # 1 = sang ph·∫£i, -1 = sang tr√°i
dy = 1  # 1 = xu·ªëng, -1 = l√™n

for _ in range(50):
    clear_output(wait=True)
    
    # V·∫Ω khung v√† b√≥ng
    for row in range(chieu_cao):
        dong = ""
        for col in range(chieu_rong):
            if row == y and col == x:
                dong = dong + "üî¥"
            else:
                dong = dong + "¬∑ "
        print(dong)
    
    # Di chuy·ªÉn b√≥ng
    x = x + dx
    y = y + dy
    
    # Ki·ªÉm tra va ch·∫°m bi√™n - ƒê·ªîI H∆Ø·ªöNG
    if x >= chieu_rong - 1 or x <= 0:
        dx = -dx  # ƒê·ªïi h∆∞·ªõng ngang
    if y >= chieu_cao - 1 or y <= 0:
        dy = -dy  # ƒê·ªïi h∆∞·ªõng d·ªçc
    
    time.sleep(0.15)

print("\n‚úÖ K·∫øt th√∫c animation!")
print("\nüí° Ki·∫øn th·ª©c: if-else ki·ªÉm tra ƒëi·ªÅu ki·ªán bi√™n")
````

### üé¨ ·ª®NG D·ª§NG TH·ª∞C T·∫æ: Countdown Timer

**Ki·∫øn th·ª©c s·ª≠ d·ª•ng:** `while` loop ƒë·ªÉ ƒë·∫øm ng∆∞·ª£c cho ƒë·∫øn khi ƒëi·ªÅu ki·ªán sai

T·∫°o ƒë·ªìng h·ªì ƒë·∫øm ng∆∞·ª£c nh∆∞ trong c√°c ·ª©ng d·ª•ng timer!

````python
# üé¨ DEMO: Countdown Timer Animation
# Ch·∫°y cell n√†y ƒë·ªÉ xem hi·ªáu ·ª©ng!

import time
from IPython.display import clear_output

# Th·ªùi gian ƒë·∫øm ng∆∞·ª£c (gi√¢y)
thoi_gian = 10

print("‚è∞ B·∫Øt ƒë·∫ßu ƒë·∫øm ng∆∞·ª£c!")
while thoi_gian > 0:
    clear_output(wait=True)
    
    # T√≠nh ph√∫t v√† gi√¢y
    phut = thoi_gian // 60
    giay = thoi_gian % 60
    
    # Hi·ªÉn th·ªã ƒë·ªìng h·ªì
    print("‚ïî‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïó")
    print(f"‚ïë   ‚è±Ô∏è  {phut:02d}:{giay:02d}    ‚ïë")
    print("‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù")
    
    thoi_gian = thoi_gian - 1  # Gi·∫£m 1 gi√¢y
    time.sleep(1)

clear_output(wait=True)
print("‚ïî‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïó")
print("‚ïë   üéâ H·∫æT GI·ªú!  ‚ïë")
print("‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù")
print("\nüí° Ki·∫øn th·ª©c: while loop + ƒëi·ªÅu ki·ªán d·ª´ng")
````

````python
# üé¨ DEMO: Text Typing Animation
# Ch·∫°y cell n√†y ƒë·ªÉ xem hi·ªáu ·ª©ng!

import time
from IPython.display import clear_output

# VƒÉn b·∫£n m·∫´u (c√≥ th·ªÉ thay b·∫±ng input() ƒë·ªÉ nh·∫≠p t√πy √Ω)
van_ban = "Xin ch√†o! ƒê√¢y l√† hi·ªáu ·ª©ng ƒë√°nh m√°y..."

print("‚å®Ô∏è Hi·ªáu ·ª©ng Typing Animation")
print("=" * 40)
print()

# Duy·ªát t·ª´ng k√Ω t·ª± trong chu·ªói
hien_thi = ""
for ky_tu in van_ban:
    hien_thi = hien_thi + ky_tu  # Th√™m k√Ω t·ª± v√†o chu·ªói hi·ªán t·∫°i
    clear_output(wait=True)
    print("‚å®Ô∏è Hi·ªáu ·ª©ng Typing Animation")
    print("=" * 40)
    print()
    print(f"üìù {hien_thi}‚ñå")  # ‚ñå l√† con tr·ªè nh·∫•p nh√°y
    time.sleep(0.08)

# K·∫øt th√∫c - x√≥a con tr·ªè
clear_output(wait=True)
print("‚å®Ô∏è Hi·ªáu ·ª©ng Typing Animation")
print("=" * 40)
print()
print(f"üìù {hien_thi}")
print()
print("‚úÖ Ho√†n th√†nh!")
print("\nüí° Ki·∫øn th·ª©c: V√≤ng l·∫∑p for duy·ªát qua chu·ªói + n·ªëi chu·ªói")
````
