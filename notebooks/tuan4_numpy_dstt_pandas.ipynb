# Section 1: PH√âP TO√ÅN MA TR·∫¨N C∆† B·∫¢N

üìå **Slide 3**

| To√°n | NumPy | √ù nghƒ©a |
|------|-------|--------|
| A + B | `A + B` | C·ªông ma tr·∫≠n |
| A √ó B | `A @ B` | Nh√¢n ma tr·∫≠n |
| A^T | `A.T` | Chuy·ªÉn v·ªã |
| c √ó A | `c * A` | Nh√¢n v√¥ h∆∞·ªõng |

**‚ö†Ô∏è Ch√∫ √Ω:** `A * B` = nh√¢n t·ª´ng ph·∫ßn t·ª≠. `A @ B` = nh√¢n ma tr·∫≠n!

````python
# === CODE M·∫™U ===
import numpy as np

A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

print("A =")
print(A)
print("\nB =")
print(B)

# C·ªông ma tr·∫≠n
print("\nA + B =")
print(A + B)

# Nh√¢n ma tr·∫≠n (KH√îNG ph·∫£i A * B!)
print("\nA @ B (nh√¢n ma tr·∫≠n) =")
print(A @ B)

print("\nA * B (nh√¢n t·ª´ng ph·∫ßn t·ª≠ - KH√ÅC!) =")
print(A * B)

# Chuy·ªÉn v·ªã
print("\nA^T =")
print(A.T)
````

# Section 4: T√çCH V√î H∆Ø·ªöNG & TR·ª∞C GIAO

üìå **Slide 6**

| C√¥ng th·ª©c | NumPy |
|-----------|-------|
| a ¬∑ b (t√≠ch v√¥ h∆∞·ªõng) | `np.dot(a, b)` |
| ‚Äña‚Äñ (chu·∫©n) | `np.linalg.norm(a)` |
| cos(Œ∏) = (a¬∑b) / (‚Äña‚Äñ¬∑‚Äñb‚Äñ) | T√≠nh t·ª´ dot v√† norm |
| a ‚ä• b (tr·ª±c giao) | `np.dot(a, b) == 0` |
| Chi·∫øu a l√™n b | `(np.dot(a,b) / np.dot(b,b)) * b` |

````python
# === CODE M·∫™U ===
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# T√≠ch v√¥ h∆∞·ªõng
dot = np.dot(a, b)  # 1*4 + 2*5 + 3*6 = 32
print(f"a ¬∑ b = {dot}")

# Chu·∫©n (ƒë·ªô d√†i)
norm_a = np.linalg.norm(a)
print(f"||a|| = {norm_a:.4f}")

# G√≥c gi·ªØa 2 vector
cos_theta = dot / (np.linalg.norm(a) * np.linalg.norm(b))
theta_rad = np.arccos(cos_theta)
theta_deg = np.degrees(theta_rad)
print(f"G√≥c gi·ªØa a v√† b = {theta_deg:.2f}¬∞")

# Ki·ªÉm tra tr·ª±c giao
u = np.array([1, 0, 0])
v = np.array([0, 1, 0])
print(f"\nu ¬∑ v = {np.dot(u, v)} ‚Üí Tr·ª±c giao!")
````

# Section 5: H·∫†NG MA TR·∫¨N & ƒê·ªòC L·∫¨P TUY·∫æN T√çNH

üìå **Slide 7**

**H·∫°ng ma tr·∫≠n** = s·ªë vector h√†ng (ho·∫∑c c·ªôt) ƒë·ªôc l·∫≠p tuy·∫øn t√≠nh t·ªëi ƒëa.

**·ª®ng d·ª•ng cho h·ªá PT Ax = b (n ·∫©n):**
- rank(A) = rank(A|b) = n ‚Üí **Nghi·ªám duy nh·∫•t**
- rank(A) = rank(A|b) < n ‚Üí **V√¥ s·ªë nghi·ªám**
- rank(A) ‚â† rank(A|b) ‚Üí **V√¥ nghi·ªám**

````python
# === CODE M·∫™U ===

# Ma tr·∫≠n rank ƒë·∫ßy ƒë·ªß
A = np.array([[1, 0, 0],
              [0, 1, 0],
              [0, 0, 1]])
print(f"rank(I) = {np.linalg.matrix_rank(A)}")  # 3

# Ma tr·∫≠n rank kh√¥ng ƒë·∫ßy ƒë·ªß (h√†ng 3 = h√†ng 1 + h√†ng 2)
B = np.array([[1, 2, 3],
              [4, 5, 6],
              [5, 7, 9]])  # h√†ng 3 = h√†ng 1 + h√†ng 2
print(f"rank(B) = {np.linalg.matrix_rank(B)}")  # 2

# Ki·ªÉm tra: h√†ng 3 c√≥ ph·∫£i t·ªï h·ª£p tuy·∫øn t√≠nh?
print(f"\nH√†ng 3 = H√†ng 1 + H√†ng 2?")
print(f"H√†ng 1 + H√†ng 2 = {B[0] + B[1]}")
print(f"H√†ng 3         = {B[2]}")
````

**B√†i to√°n tr·ªã ri√™ng:** T√¨m Œª (tr·ªã ri√™ng) v√† v (vector ri√™ng) sao cho **Av = Œªv**

- Vector ri√™ng v: h∆∞·ªõng kh√¥ng ƒë·ªïi khi nh√¢n v·ªõi A
- Tr·ªã ri√™ng Œª: h·ªá s·ªë co gi√£n theo h∆∞·ªõng ƒë√≥

```python
eigenvalues, eigenvectors = np.linalg.eig(A)
# eigenvalues[i] ‚Üí tr·ªã ri√™ng th·ª© i
# eigenvectors[:, i] ‚Üí vector ri√™ng t∆∞∆°ng ·ª©ng (c·ªôt th·ª© i)
```

````python
# === CODE M·∫™U ===
A = np.array([[4, 1],
              [2, 3]])

# T√¨m tr·ªã ri√™ng v√† vector ri√™ng
eigenvalues, eigenvectors = np.linalg.eig(A)

print("Ma tr·∫≠n A:")
print(A)
print(f"\nTr·ªã ri√™ng: {eigenvalues}")
print(f"\nVector ri√™ng (m·ªói C·ªòT l√† 1 vector):")
print(eigenvectors)

# Ki·ªÉm tra: Av = Œªv
for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    v = eigenvectors[:, i]  # L·∫•y C·ªòT th·ª© i
    
    Av = A @ v
    lam_v = lam * v
    
    print(f"\nŒª{i+1} = {lam:.2f}, v{i+1} = {np.round(v, 4)}")
    print(f"  Av   = {np.round(Av, 4)}")
    print(f"  Œªv   = {np.round(lam_v, 4)}")
    print(f"  Av = Œªv ? {np.allclose(Av, lam_v)}")
````
